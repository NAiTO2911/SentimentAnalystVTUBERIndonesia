```python
import os, re, json, string, pickle

import nltk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# nltk.download('punkt')
```

```python
nltk.download('stopwords')
#stopword removal and lemmatization
stopwords = nltk.corpus.stopwords.words('indonesian')
```

```python
# Create Sastrawi stemmer
stemmer = StemmerFactory().create_stemmer()
```

```python
# Tokenizer
def tokenizer(text):
    words = word_tokenize(text)
    tokens=[]
    for w in words:
        # add tokens
        if len(w) > 3 and w not in stopwords:
            w = stemmer.stem(w)
            tokens.append(w.lower())
    return " ".join(tokens)
```

```python
text=[" canggung kenapa takut kakak tiba nyulik aku kah ",
" apa aja bebas yang penting enak ",
" waduh abang abangan nih btw andi adek nya banyak ya wkwkw ",
" ini kak ",
" astaga kalian berdua aowkkwkw ",
" gila jam belum tidur anda pekerja keras sekali ya termasuk yg komen juga warbyasah ",
" ya ampun ",
" lautan asin bisa kekurangan gula awokawok ",
" ga ngerti tolong di jelaskan ",
" ari si anak tersesat "
]
```

```python
dict = {'Content':text}
df = pd.DataFrame(dict)

df['Content'] = df['Content'].apply(lambda x : tokenizer(x))
df
```

```python
#using the count vectorizer
count = CountVectorizer()
word_count=count.fit_transform(text)
print(word_count)
```

```python
word_count.shape
```

```python
print(word_count.toarray())
```

```python
tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(word_count)
df_idf = pd.DataFrame(tfidf_transformer.idf_, index=count.get_feature_names_out(),columns=["idf_weights"])
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
#inverse document frequency
dfi = df_idf.sort_values(by=['idf_weights'])
dfi
# dfi.to_clipboard(sep=',')
```

```python
#tfidf
tf_idf_vector=tfidf_transformer.transform(word_count)
feature_names = count.get_feature_names_out()
```

```python
first_document_vector=tf_idf_vector[1]
df_tfifd= pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
```

```python
tfidf = df_tfifd.sort_values(by=["tfidf"],ascending=False)
tfidf
# tfidf.to_clipboard(sep=',')
```

```python
# Fungsi save hasil data yang sudah dibersihkan
tfidf.to_csv('tfidf.csv')
```
